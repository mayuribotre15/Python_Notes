```python
#Q6. Use a dictionary comprehension to count the length of each word in a sentence
```

```python
string = 'hello to this new world'
words = string.split(" ")
count_len_word = {x:len(x) for x in words}
count_len_word
```

```python
#Q8. Use a nested list comprehension to find all of the numbers from 1â€“1000 
#that are divisible by any single digit besides 1 (2â€“9)
```

```python
nums = list(range(1,1001))
div = list(range(2,10))
num_div = [i for i in nums if True in [True for div in range(2,10) if i%div == 0]]
num_div
```
